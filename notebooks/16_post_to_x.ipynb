# Post to X (@quaksai) — Test Notebook

Tests the `post_to_x` function from `app/utils/data_ingestion_utils.py`.

**Prerequisites:** Set the following environment variables in your `.env` file:
- `X_CONSUMER_KEY` — X OAuth 1.0a API Key
- `X_CONSUMER_SECRET` — X OAuth 1.0a API Key Secret
- `X_ACCESS_TOKEN` — X OAuth 1.0a Access Token
- `X_ACCESS_TOKEN_SECRET` — X OAuth 1.0a Access Token Secret

```python
import os
from dotenv import load_dotenv

os.chdir("..")
load_dotenv()

X_CONSUMER_KEY = os.getenv("X_CONSUMER_KEY")
X_CONSUMER_SECRET = os.getenv("X_CONSUMER_SECRET")
X_ACCESS_TOKEN = os.getenv("X_ACCESS_TOKEN")
X_ACCESS_TOKEN_SECRET = os.getenv("X_ACCESS_TOKEN_SECRET")
```

### Test 1: Post with sample executive summary

```python
from app.utils.data_ingestion_utils import post_to_x

sample_summary = "Markets rally on Iran ceasefire hopes as AI chipmakers surge and crypto steadies, but regulatory risks loom over Super Micro and stablecoin yields."
sample_url = "https://quaks.ai/insights/news/item/quaks_insights-news_latest/fceb5f8fa30fb7c9f69c3ebd4b92498e"

result = post_to_x(
    executive_summary=sample_summary,
    article_url=sample_url,
    consumer_key=X_CONSUMER_KEY,
    consumer_secret=X_CONSUMER_SECRET,
    access_token=X_ACCESS_TOKEN,
    access_token_secret=X_ACCESS_TOKEN_SECRET,
)

print(f"Result: {result}")
```

### Test 2: Verify empty summary is skipped

```python
result_empty = post_to_x(
    executive_summary="",
    article_url=sample_url,
    consumer_key=X_CONSUMER_KEY,
    consumer_secret=X_CONSUMER_SECRET,
    access_token=X_ACCESS_TOKEN,
    access_token_secret=X_ACCESS_TOKEN_SECRET,
)

assert result_empty is None, "Expected None for empty summary"
print("Empty summary correctly skipped.")
```
