```python
import os
import requests
from dotenv import load_dotenv

os.chdir("..")
load_dotenv()

from app.utils.data_ingestion_utils import (
    ingest_stocks_eod,
    ingest_stocks_insider_trades,
    ingest_stocks_metadata,
    ingest_stocks_fundamental_income_statement,
    ingest_stocks_fundamental_balance_sheet,
    ingest_stocks_fundamental_cash_flow,
    ingest_stocks_fundamental_earnings_estimates,
    ingest_markets_news
)
```

```python
# api_endpoint = "https://quaks.ai"
# indexed_key_ticker_list = requests.get(f"{api_endpoint}/json/indexed_key_ticker_list.json").json()

indexed_key_ticker_list = [{
    "key_ticker": "NVDA",
    "index": "nasdaq_100",
    "name": "NVIDIA Corporation"
}]

for company in indexed_key_ticker_list:
    # ohlcv
    stocks_eod_response = ingest_stocks_eod(company["key_ticker"], index_suffix=company["index"])
    print(f"Ingestion complete stocks EOD for {company}, errors: {stocks_eod_response.json().get('errors')}")

    # insider trades
    stocks_insider_trades_response = ingest_stocks_insider_trades(company["key_ticker"], index_suffix=company["index"])
    print(f"Stocks insider trades ingestion complete with errors: {stocks_insider_trades_response.json().get('errors')}")

    # company metadata
    stocks_metadata_response = ingest_stocks_metadata(company["key_ticker"], index_suffix=company["index"])
    print(f"Stocks metadata ingestion complete with errors: {stocks_metadata_response.json().get('errors')}")

    # income statements
    stocks_fundamental_income_statement_response = ingest_stocks_fundamental_income_statement(company["key_ticker"], index_suffix=company["index"])
    print(f"Stocks fundamental income statement ingestion complete with errors: {stocks_fundamental_income_statement_response.json().get('errors')}")

    # balance sheets
    stocks_fundamental_balance_sheet_response = ingest_stocks_fundamental_balance_sheet(company["key_ticker"], index_suffix=company["index"])
    print(f"Stocks fundamental balance sheet ingestion complete with errors: {stocks_fundamental_balance_sheet_response.json().get('errors')}")

    # cash flow
    stocks_fundamental_cash_flow_response = ingest_stocks_fundamental_cash_flow(company["key_ticker"], index_suffix=company["index"])
    print(f"Stocks fundamental cash flow ingestion complete with errors: {stocks_fundamental_cash_flow_response.json().get('errors')}")

    # estimated earnings
    stocks_fundamental_earnings_estimates_response = ingest_stocks_fundamental_earnings_estimates(company["key_ticker"], index_suffix=company["index"])
    print(f"Stocks fundamental estimated earnings ingestion complete with errors: {stocks_fundamental_earnings_estimates_response.json().get('errors')}")

    # news
    markets_news_response = ingest_markets_news(company["key_ticker"], index_suffix=company["index"])
    print(f"Markets news ingestion complete with errors: {markets_news_response.json().get('errors')}")
```
